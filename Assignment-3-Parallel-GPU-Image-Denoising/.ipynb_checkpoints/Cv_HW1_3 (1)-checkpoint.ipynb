```python
# coding=utf-8
!pip install pycuda
import time
import cv2
import numpy as np
import pycuda.driver as drv
import pycuda.tools
import pycuda.autoinit
from pycuda.compiler import SourceModule
import pycuda.gpuarray as gpuarray
import pycuda.cumath
from pycuda.elementwise import ElementwiseKernel
from pycuda.compiler import SourceModule



# Load the image (grayscale or color as needed)
img = cv2.imread('bird.jpg', cv2.IMREAD_COLOR)  # or cv2.IMREAD_COLOR for color images

# img = cv2.resize(img, (16, 16), interpolation=cv2.INTER_AREA)

# Parameters for Gaussian noise
mean = 0
stddev = 20  # Standard deviation for noise level

# Generate Gaussian noise
gaussian_noise = np.random.normal(mean, stddev, img.shape).astype(np.float32)

# Add the Gaussian noise to the image
noisy_img = img.astype(np.float32) + gaussian_noise

# Clip values to stay within the valid range [0, 255]
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

# Save the noisy image
cv2.imwrite('noisy_bird.jpg', noisy_img)

# Display the noisy image (optional)
# cv2.imshow('Noisy Image', noisy_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Fast NLM Denoise filter for monochrome image V1.1
# Based on Fast NLM filter CUDA program provided by Nvidia - CUDA SDK samples
# Alain PAILLOU - August 2019


In_File = 'noisy_bird.jpg'
Out_File_OpenCV = "opencv_cpu_bird.jpg"
Out_File_PyCuda = "pycu_gpu_bird.jpg"




mod = SourceModule("""
__global__ void NLM2_Colour(unsigned char *dest_r, unsigned char *dest_g, unsigned char *dest_b, unsigned char *img_r, unsigned char *img_g,
unsigned char *img_b, int imageW, int imageH, float Noise, float lerpC)
{

    #define NLM_WINDOW_RADIUS   3
    #define NLM_BLOCK_RADIUS    3

    #define NLM_WEIGHT_THRESHOLD    0.00039f
    #define NLM_LERP_THRESHOLD      0.10f

    __shared__ float fWeights[64];

    const float NLM_WINDOW_AREA = (2.0 * NLM_WINDOW_RADIUS + 1.0) * (2.0 * NLM_WINDOW_RADIUS + 1.0) ;
    const float INV_NLM_WINDOW_AREA = (1.0 / NLM_WINDOW_AREA);

    const long int   ix = blockDim.x * blockIdx.x + threadIdx.x;
    const long int   iy = blockDim.y * blockIdx.y + threadIdx.y;

    const float  x = (float)ix  + 1.0f;
    const float  y = (float)iy  + 1.0f;
    const float cx = blockDim.x * blockIdx.x + NLM_WINDOW_RADIUS + 1.0f;
    const float cy = blockDim.x * blockIdx.y + NLM_WINDOW_RADIUS + 1.0f;
    const float limxmin = NLM_BLOCK_RADIUS + 2;
    const float limxmax = imageW - NLM_BLOCK_RADIUS - 2;
    const float limymin = NLM_BLOCK_RADIUS + 2;
    const float limymax = imageH - NLM_BLOCK_RADIUS - 2;


    long int index4;
    long int index5;

    if(ix>limxmin && ix<limxmax && iy>limymin && iy<limymax){
        //Find color distance from current texel to the center of NLM window
        float weight = 0;

        for(float n = -NLM_BLOCK_RADIUS; n <= NLM_BLOCK_RADIUS; n++)
            for(float m = -NLM_BLOCK_RADIUS; m <= NLM_BLOCK_RADIUS; m++) {
                long int index1 = cx + m + (cy + n) * imageW;
                long int index2 = x + m + (y + n) * imageW;
                weight += ((img_r[index2] - img_r[index1]) * (img_r[index2] - img_r[index1])
                + (img_g[index2] - img_g[index1]) * (img_g[index2] - img_g[index1])
                + (img_b[index2] - img_b[index1]) * (img_b[index2] - img_b[index1])) / (256.0 * 256.0);
                }

        //Geometric distance from current texel to the center of NLM window
        float dist =
            (threadIdx.x - NLM_WINDOW_RADIUS) * (threadIdx.x - NLM_WINDOW_RADIUS) +
            (threadIdx.y - NLM_WINDOW_RADIUS) * (threadIdx.y - NLM_WINDOW_RADIUS);

        //Derive final weight from color and geometric distance
        weight = __expf(-(weight * Noise + dist * INV_NLM_WINDOW_AREA));

        //Write the result to shared memory
        fWeights[threadIdx.y * 8 + threadIdx.x] = weight / 256.0;
        //Wait until all the weights are ready
        __syncthreads();


        //Normalized counter for the NLM weight threshold
        float fCount = 0;
        //Total sum of pixel weights
        float sumWeights = 0;
        //Result accumulator
        float3 clr = {0.0, 0.0, 0.0};

        int idx = 0;

        //Cycle through NLM window, surrounding (x, y) texel

        for(float i = -NLM_WINDOW_RADIUS; i <= NLM_WINDOW_RADIUS + 1; i++)
            for(float j = -NLM_WINDOW_RADIUS; j <= NLM_WINDOW_RADIUS + 1; j++)
            {
                //Load precomputed weight
                float weightIJ = fWeights[idx++];

                //Accumulate (x + j, y + i) texel color with computed weight
                float3 clrIJ ; // Ligne code modifiÃ©e
                int index3 = x + j + (y + i) * imageW;
                clrIJ.x = img_r[index3];
                clrIJ.y = img_g[index3];
                clrIJ.z = img_b[index3];

                clr.x += clrIJ.x * weightIJ;
                clr.y += clrIJ.y * weightIJ;
                clr.z += clrIJ.z * weightIJ;

                //Sum of weights for color normalization to [0..1] range
                sumWeights  += weightIJ;

                //Update weight counter, if NLM weight for current window texel
                //exceeds the weight threshold
                fCount += (weightIJ > NLM_WEIGHT_THRESHOLD) ? INV_NLM_WINDOW_AREA : 0;
            }

        //Normalize result color by sum of weights
        sumWeights = 0.0039f / sumWeights;
        clr.x *= sumWeights;
        clr.y *= sumWeights;
        clr.z *= sumWeights;

        //Choose LERP quotent basing on how many texels
        //within the NLM window exceeded the weight threshold
        float lerpQ = (fCount > NLM_LERP_THRESHOLD) ? lerpC : 1.0f - lerpC;

        //Write final result to global memory
        float3 clr00 = {0.0, 0.0, 0.0};
        index4 = x + (y * imageW);
        index5 = imageW * (iy + 1) + ix + 1;

        clr00.x = img_r[index4] / 256.0;
        clr00.y = img_g[index4] / 256.0;
        clr00.z = img_b[index4] / 256.0;

        clr.x = clr.x + (clr00.x - clr.x) * lerpQ;
        clr.y = clr.y + (clr00.y - clr.y) * lerpQ;
        clr.z = clr.z + (clr00.z - clr.z) * lerpQ;


        dest_r[index5] = (int)(clr.x * 256.0);
        dest_g[index5] = (int)(clr.y * 256.0);
        dest_b[index5] = (int)(clr.z * 256.0);
    }
}
""")

NLM2_Colour_GPU = mod.get_function("NLM2_Colour")


image_brut_CV = cv2.imread(In_File,cv2.IMREAD_COLOR)
height,width,layers = image_brut_CV.shape
nb_pixels = height * width

# Determination taille blocks et Grid
nb_ThreadsX = 8
nb_ThreadsY = 8
nb_blocksX = (width // nb_ThreadsX) + 1
nb_blocksY = (height // nb_ThreadsY) + 1

# Set NLM parameters
NLM_Noise = 1.45
Noise = 1.0/(NLM_Noise*NLM_Noise)
lerpC = 0.2


# Algorithm CPU using OpenCV fastNlMeansDenoisingColored routine
tps1 = time.time()
param=21.0
image_brut_CPU=cv2.fastNlMeansDenoisingColored(image_brut_CV, None, param, param, 3, 5) # application filtre denoise software colour
image_brut_CPU=cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB)
tps_CPU_bird = time.time() - tps1
print("OpenCV treatment OK")
print ("CPU treatment time : ",tps_CPU_bird)
print("")
cv2.imwrite(Out_File_OpenCV, cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


# Algorithm GPU using PyCuda
print("Test GPU ",nb_blocksX*nb_blocksY," Blocks ",nb_ThreadsX*nb_ThreadsY," Threads/Block")
tps1 = time.time()

b,g,r = cv2.split(image_brut_CV)
b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
img_b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
drv.memcpy_htod(b_gpu, b)
drv.memcpy_htod(img_b_gpu, b)
res_b = np.empty_like(b)

g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(g_gpu, g)
img_g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(img_g_gpu, g)
res_g = np.empty_like(g)

r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(r_gpu, r)
img_r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(img_r_gpu, r)
res_r = np.empty_like(r)

NLM2_Colour_GPU(r_gpu, g_gpu, b_gpu, img_r_gpu, img_g_gpu, img_b_gpu,np.intc(width),np.intc(height), np.float32(Noise), \
         np.float32(lerpC), block=(nb_ThreadsX,nb_ThreadsY,1), grid=(nb_blocksX,nb_blocksY))

drv.memcpy_dtoh(res_r, r_gpu)
drv.memcpy_dtoh(res_g, g_gpu)
drv.memcpy_dtoh(res_b, b_gpu)
r_gpu.free()
g_gpu.free()
b_gpu.free()
img_r_gpu.free()
img_g_gpu.free()
img_b_gpu.free()

image_brut_GPU=cv2.merge((res_r,res_g,res_b))

tps_GPU_bird = time.time() - tps1

print("PyCuda treatment OK")
print ("GPU treatment time : ",tps_GPU_bird)
print("")

cv2.imwrite(Out_File_PyCuda, cv2.cvtColor(image_brut_GPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


Ecart1_bird = tps_CPU_bird/tps_GPU_bird
print ("Acceleration factor CPU/GPU : ",Ecart1_bird)
```

```python
# coding=utf-8
!pip install pycuda
import time
import cv2
import numpy as np
import pycuda.driver as drv
import pycuda.tools
import pycuda.autoinit
from pycuda.compiler import SourceModule
import pycuda.gpuarray as gpuarray
import pycuda.cumath
from pycuda.elementwise import ElementwiseKernel
from pycuda.compiler import SourceModule



# Load the image (grayscale or color as needed)
img = cv2.imread('woman.jpg', cv2.IMREAD_COLOR)  # or cv2.IMREAD_COLOR for color images

# img = cv2.resize(img, (16, 16), interpolation=cv2.INTER_AREA)

# Parameters for Gaussian noise
mean = 0
stddev = 20  # Standard deviation for noise level

# Generate Gaussian noise
gaussian_noise = np.random.normal(mean, stddev, img.shape).astype(np.float32)

# Add the Gaussian noise to the image
noisy_img = img.astype(np.float32) + gaussian_noise

# Clip values to stay within the valid range [0, 255]
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

# Save the noisy image
cv2.imwrite('noisy_woman.jpg', noisy_img)

# Display the noisy image (optional)
# cv2.imshow('Noisy Image', noisy_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Fast NLM Denoise filter for monochrome image V1.1
# Based on Fast NLM filter CUDA program provided by Nvidia - CUDA SDK samples
# Alain PAILLOU - August 2019


In_File = 'noisy_woman.jpg'
Out_File_OpenCV = "opencv_cpu_woman.jpg"
Out_File_PyCuda = "pycu_gpu_woman.jpg"




mod = SourceModule("""
__global__ void NLM2_Colour(unsigned char *dest_r, unsigned char *dest_g, unsigned char *dest_b, unsigned char *img_r, unsigned char *img_g,
unsigned char *img_b, int imageW, int imageH, float Noise, float lerpC)
{

    #define NLM_WINDOW_RADIUS   3
    #define NLM_BLOCK_RADIUS    3

    #define NLM_WEIGHT_THRESHOLD    0.00039f
    #define NLM_LERP_THRESHOLD      0.10f

    __shared__ float fWeights[64];

    const float NLM_WINDOW_AREA = (2.0 * NLM_WINDOW_RADIUS + 1.0) * (2.0 * NLM_WINDOW_RADIUS + 1.0) ;
    const float INV_NLM_WINDOW_AREA = (1.0 / NLM_WINDOW_AREA);

    const long int   ix = blockDim.x * blockIdx.x + threadIdx.x;
    const long int   iy = blockDim.y * blockIdx.y + threadIdx.y;

    const float  x = (float)ix  + 1.0f;
    const float  y = (float)iy  + 1.0f;
    const float cx = blockDim.x * blockIdx.x + NLM_WINDOW_RADIUS + 1.0f;
    const float cy = blockDim.x * blockIdx.y + NLM_WINDOW_RADIUS + 1.0f;
    const float limxmin = NLM_BLOCK_RADIUS + 2;
    const float limxmax = imageW - NLM_BLOCK_RADIUS - 2;
    const float limymin = NLM_BLOCK_RADIUS + 2;
    const float limymax = imageH - NLM_BLOCK_RADIUS - 2;


    long int index4;
    long int index5;

    if(ix>limxmin && ix<limxmax && iy>limymin && iy<limymax){
        //Find color distance from current texel to the center of NLM window
        float weight = 0;

        for(float n = -NLM_BLOCK_RADIUS; n <= NLM_BLOCK_RADIUS; n++)
            for(float m = -NLM_BLOCK_RADIUS; m <= NLM_BLOCK_RADIUS; m++) {
                long int index1 = cx + m + (cy + n) * imageW;
                long int index2 = x + m + (y + n) * imageW;
                weight += ((img_r[index2] - img_r[index1]) * (img_r[index2] - img_r[index1])
                + (img_g[index2] - img_g[index1]) * (img_g[index2] - img_g[index1])
                + (img_b[index2] - img_b[index1]) * (img_b[index2] - img_b[index1])) / (256.0 * 256.0);
                }

        //Geometric distance from current texel to the center of NLM window
        float dist =
            (threadIdx.x - NLM_WINDOW_RADIUS) * (threadIdx.x - NLM_WINDOW_RADIUS) +
            (threadIdx.y - NLM_WINDOW_RADIUS) * (threadIdx.y - NLM_WINDOW_RADIUS);

        //Derive final weight from color and geometric distance
        weight = __expf(-(weight * Noise + dist * INV_NLM_WINDOW_AREA));

        //Write the result to shared memory
        fWeights[threadIdx.y * 8 + threadIdx.x] = weight / 256.0;
        //Wait until all the weights are ready
        __syncthreads();


        //Normalized counter for the NLM weight threshold
        float fCount = 0;
        //Total sum of pixel weights
        float sumWeights = 0;
        //Result accumulator
        float3 clr = {0.0, 0.0, 0.0};

        int idx = 0;

        //Cycle through NLM window, surrounding (x, y) texel

        for(float i = -NLM_WINDOW_RADIUS; i <= NLM_WINDOW_RADIUS + 1; i++)
            for(float j = -NLM_WINDOW_RADIUS; j <= NLM_WINDOW_RADIUS + 1; j++)
            {
                //Load precomputed weight
                float weightIJ = fWeights[idx++];

                //Accumulate (x + j, y + i) texel color with computed weight
                float3 clrIJ ; // Ligne code modifiÃ©e
                int index3 = x + j + (y + i) * imageW;
                clrIJ.x = img_r[index3];
                clrIJ.y = img_g[index3];
                clrIJ.z = img_b[index3];

                clr.x += clrIJ.x * weightIJ;
                clr.y += clrIJ.y * weightIJ;
                clr.z += clrIJ.z * weightIJ;

                //Sum of weights for color normalization to [0..1] range
                sumWeights  += weightIJ;

                //Update weight counter, if NLM weight for current window texel
                //exceeds the weight threshold
                fCount += (weightIJ > NLM_WEIGHT_THRESHOLD) ? INV_NLM_WINDOW_AREA : 0;
            }

        //Normalize result color by sum of weights
        sumWeights = 0.0039f / sumWeights;
        clr.x *= sumWeights;
        clr.y *= sumWeights;
        clr.z *= sumWeights;

        //Choose LERP quotent basing on how many texels
        //within the NLM window exceeded the weight threshold
        float lerpQ = (fCount > NLM_LERP_THRESHOLD) ? lerpC : 1.0f - lerpC;

        //Write final result to global memory
        float3 clr00 = {0.0, 0.0, 0.0};
        index4 = x + (y * imageW);
        index5 = imageW * (iy + 1) + ix + 1;

        clr00.x = img_r[index4] / 256.0;
        clr00.y = img_g[index4] / 256.0;
        clr00.z = img_b[index4] / 256.0;

        clr.x = clr.x + (clr00.x - clr.x) * lerpQ;
        clr.y = clr.y + (clr00.y - clr.y) * lerpQ;
        clr.z = clr.z + (clr00.z - clr.z) * lerpQ;


        dest_r[index5] = (int)(clr.x * 256.0);
        dest_g[index5] = (int)(clr.y * 256.0);
        dest_b[index5] = (int)(clr.z * 256.0);
    }
}
""")

NLM2_Colour_GPU = mod.get_function("NLM2_Colour")


image_brut_CV = cv2.imread(In_File,cv2.IMREAD_COLOR)
height,width,layers = image_brut_CV.shape
nb_pixels = height * width

# Determination taille blocks et Grid
nb_ThreadsX = 8
nb_ThreadsY = 8
nb_blocksX = (width // nb_ThreadsX) + 1
nb_blocksY = (height // nb_ThreadsY) + 1

# Set NLM parameters
NLM_Noise = 1.45
Noise = 1.0/(NLM_Noise*NLM_Noise)
lerpC = 0.2


# Algorithm CPU using OpenCV fastNlMeansDenoisingColored routine
tps1 = time.time()
param=21.0
image_brut_CPU=cv2.fastNlMeansDenoisingColored(image_brut_CV, None, param, param, 3, 5) # application filtre denoise software colour
image_brut_CPU=cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB)
tps_CPU_woman = time.time() - tps1
print("OpenCV treatment OK")
print ("CPU treatment time : ",tps_CPU_woman)
print("")
cv2.imwrite(Out_File_OpenCV, cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


# Algorithm GPU using PyCuda
print("Test GPU ",nb_blocksX*nb_blocksY," Blocks ",nb_ThreadsX*nb_ThreadsY," Threads/Block")
tps1 = time.time()

b,g,r = cv2.split(image_brut_CV)
b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
img_b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
drv.memcpy_htod(b_gpu, b)
drv.memcpy_htod(img_b_gpu, b)
res_b = np.empty_like(b)

g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(g_gpu, g)
img_g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(img_g_gpu, g)
res_g = np.empty_like(g)

r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(r_gpu, r)
img_r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(img_r_gpu, r)
res_r = np.empty_like(r)

NLM2_Colour_GPU(r_gpu, g_gpu, b_gpu, img_r_gpu, img_g_gpu, img_b_gpu,np.intc(width),np.intc(height), np.float32(Noise), \
         np.float32(lerpC), block=(nb_ThreadsX,nb_ThreadsY,1), grid=(nb_blocksX,nb_blocksY))

drv.memcpy_dtoh(res_r, r_gpu)
drv.memcpy_dtoh(res_g, g_gpu)
drv.memcpy_dtoh(res_b, b_gpu)
r_gpu.free()
g_gpu.free()
b_gpu.free()
img_r_gpu.free()
img_g_gpu.free()
img_b_gpu.free()

image_brut_GPU=cv2.merge((res_r,res_g,res_b))

tps_GPU_woman = time.time() - tps1

print("PyCuda treatment OK")
print ("GPU treatment time : ",tps_GPU_woman)
print("")

cv2.imwrite(Out_File_PyCuda, cv2.cvtColor(image_brut_GPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


Ecart1_woman = tps_CPU_woman/tps_GPU_woman
print ("Acceleration factor CPU/GPU : ",Ecart1_woman)
```

```python
# coding=utf-8
!pip install pycuda
import time
import cv2
import numpy as np
import pycuda.driver as drv
import pycuda.tools
import pycuda.autoinit
from pycuda.compiler import SourceModule
import pycuda.gpuarray as gpuarray
import pycuda.cumath
from pycuda.elementwise import ElementwiseKernel
from pycuda.compiler import SourceModule



# Load the image (grayscale or color as needed)
img = cv2.imread('vegetables.jpg', cv2.IMREAD_COLOR)  # or cv2.IMREAD_COLOR for color images

# img = cv2.resize(img, (16, 16), interpolation=cv2.INTER_AREA)

# Parameters for Gaussian noise
mean = 0
stddev = 20  # Standard deviation for noise level

# Generate Gaussian noise
gaussian_noise = np.random.normal(mean, stddev, img.shape).astype(np.float32)

# Add the Gaussian noise to the image
noisy_img = img.astype(np.float32) + gaussian_noise

# Clip values to stay within the valid range [0, 255]
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

# Save the noisy image
cv2.imwrite('noisy_vegetables.jpg', noisy_img)

# Display the noisy image (optional)
# cv2.imshow('Noisy Image', noisy_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Fast NLM Denoise filter for monochrome image V1.1
# Based on Fast NLM filter CUDA program provided by Nvidia - CUDA SDK samples
# Alain PAILLOU - August 2019


In_File = 'noisy_vegetables.jpg'
Out_File_OpenCV = "opencv_cpu_vegetables.jpg"
Out_File_PyCuda = "pycu_gpu_vegetables.jpg"




mod = SourceModule("""
__global__ void NLM2_Colour(unsigned char *dest_r, unsigned char *dest_g, unsigned char *dest_b, unsigned char *img_r, unsigned char *img_g,
unsigned char *img_b, int imageW, int imageH, float Noise, float lerpC)
{

    #define NLM_WINDOW_RADIUS   3
    #define NLM_BLOCK_RADIUS    3

    #define NLM_WEIGHT_THRESHOLD    0.00039f
    #define NLM_LERP_THRESHOLD      0.10f

    __shared__ float fWeights[64];

    const float NLM_WINDOW_AREA = (2.0 * NLM_WINDOW_RADIUS + 1.0) * (2.0 * NLM_WINDOW_RADIUS + 1.0) ;
    const float INV_NLM_WINDOW_AREA = (1.0 / NLM_WINDOW_AREA);

    const long int   ix = blockDim.x * blockIdx.x + threadIdx.x;
    const long int   iy = blockDim.y * blockIdx.y + threadIdx.y;

    const float  x = (float)ix  + 1.0f;
    const float  y = (float)iy  + 1.0f;
    const float cx = blockDim.x * blockIdx.x + NLM_WINDOW_RADIUS + 1.0f;
    const float cy = blockDim.x * blockIdx.y + NLM_WINDOW_RADIUS + 1.0f;
    const float limxmin = NLM_BLOCK_RADIUS + 2;
    const float limxmax = imageW - NLM_BLOCK_RADIUS - 2;
    const float limymin = NLM_BLOCK_RADIUS + 2;
    const float limymax = imageH - NLM_BLOCK_RADIUS - 2;


    long int index4;
    long int index5;

    if(ix>limxmin && ix<limxmax && iy>limymin && iy<limymax){
        //Find color distance from current texel to the center of NLM window
        float weight = 0;

        for(float n = -NLM_BLOCK_RADIUS; n <= NLM_BLOCK_RADIUS; n++)
            for(float m = -NLM_BLOCK_RADIUS; m <= NLM_BLOCK_RADIUS; m++) {
                long int index1 = cx + m + (cy + n) * imageW;
                long int index2 = x + m + (y + n) * imageW;
                weight += ((img_r[index2] - img_r[index1]) * (img_r[index2] - img_r[index1])
                + (img_g[index2] - img_g[index1]) * (img_g[index2] - img_g[index1])
                + (img_b[index2] - img_b[index1]) * (img_b[index2] - img_b[index1])) / (256.0 * 256.0);
                }

        //Geometric distance from current texel to the center of NLM window
        float dist =
            (threadIdx.x - NLM_WINDOW_RADIUS) * (threadIdx.x - NLM_WINDOW_RADIUS) +
            (threadIdx.y - NLM_WINDOW_RADIUS) * (threadIdx.y - NLM_WINDOW_RADIUS);

        //Derive final weight from color and geometric distance
        weight = __expf(-(weight * Noise + dist * INV_NLM_WINDOW_AREA));

        //Write the result to shared memory
        fWeights[threadIdx.y * 8 + threadIdx.x] = weight / 256.0;
        //Wait until all the weights are ready
        __syncthreads();


        //Normalized counter for the NLM weight threshold
        float fCount = 0;
        //Total sum of pixel weights
        float sumWeights = 0;
        //Result accumulator
        float3 clr = {0.0, 0.0, 0.0};

        int idx = 0;

        //Cycle through NLM window, surrounding (x, y) texel

        for(float i = -NLM_WINDOW_RADIUS; i <= NLM_WINDOW_RADIUS + 1; i++)
            for(float j = -NLM_WINDOW_RADIUS; j <= NLM_WINDOW_RADIUS + 1; j++)
            {
                //Load precomputed weight
                float weightIJ = fWeights[idx++];

                //Accumulate (x + j, y + i) texel color with computed weight
                float3 clrIJ ; // Ligne code modifiÃ©e
                int index3 = x + j + (y + i) * imageW;
                clrIJ.x = img_r[index3];
                clrIJ.y = img_g[index3];
                clrIJ.z = img_b[index3];

                clr.x += clrIJ.x * weightIJ;
                clr.y += clrIJ.y * weightIJ;
                clr.z += clrIJ.z * weightIJ;

                //Sum of weights for color normalization to [0..1] range
                sumWeights  += weightIJ;

                //Update weight counter, if NLM weight for current window texel
                //exceeds the weight threshold
                fCount += (weightIJ > NLM_WEIGHT_THRESHOLD) ? INV_NLM_WINDOW_AREA : 0;
            }

        //Normalize result color by sum of weights
        sumWeights = 0.0039f / sumWeights;
        clr.x *= sumWeights;
        clr.y *= sumWeights;
        clr.z *= sumWeights;

        //Choose LERP quotent basing on how many texels
        //within the NLM window exceeded the weight threshold
        float lerpQ = (fCount > NLM_LERP_THRESHOLD) ? lerpC : 1.0f - lerpC;

        //Write final result to global memory
        float3 clr00 = {0.0, 0.0, 0.0};
        index4 = x + (y * imageW);
        index5 = imageW * (iy + 1) + ix + 1;

        clr00.x = img_r[index4] / 256.0;
        clr00.y = img_g[index4] / 256.0;
        clr00.z = img_b[index4] / 256.0;

        clr.x = clr.x + (clr00.x - clr.x) * lerpQ;
        clr.y = clr.y + (clr00.y - clr.y) * lerpQ;
        clr.z = clr.z + (clr00.z - clr.z) * lerpQ;


        dest_r[index5] = (int)(clr.x * 256.0);
        dest_g[index5] = (int)(clr.y * 256.0);
        dest_b[index5] = (int)(clr.z * 256.0);
    }
}
""")

NLM2_Colour_GPU = mod.get_function("NLM2_Colour")


image_brut_CV = cv2.imread(In_File,cv2.IMREAD_COLOR)
height,width,layers = image_brut_CV.shape
nb_pixels = height * width

# Determination taille blocks et Grid
nb_ThreadsX = 8
nb_ThreadsY = 8
nb_blocksX = (width // nb_ThreadsX) + 1
nb_blocksY = (height // nb_ThreadsY) + 1

# Set NLM parameters
NLM_Noise = 1.45
Noise = 1.0/(NLM_Noise*NLM_Noise)
lerpC = 0.2


# Algorithm CPU using OpenCV fastNlMeansDenoisingColored routine
tps1 = time.time()
param=21.0
image_brut_CPU=cv2.fastNlMeansDenoisingColored(image_brut_CV, None, param, param, 3, 5) # application filtre denoise software colour
image_brut_CPU=cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB)
tps_CPU_vegetables = time.time() - tps1
print("OpenCV treatment OK")
print ("CPU treatment time : ",tps_CPU_vegetables)
print("")
cv2.imwrite(Out_File_OpenCV, cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


# Algorithm GPU using PyCuda
print("Test GPU ",nb_blocksX*nb_blocksY," Blocks ",nb_ThreadsX*nb_ThreadsY," Threads/Block")
tps1 = time.time()

b,g,r = cv2.split(image_brut_CV)
b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
img_b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
drv.memcpy_htod(b_gpu, b)
drv.memcpy_htod(img_b_gpu, b)
res_b = np.empty_like(b)

g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(g_gpu, g)
img_g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
drv.memcpy_htod(img_g_gpu, g)
res_g = np.empty_like(g)

r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(r_gpu, r)
img_r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
drv.memcpy_htod(img_r_gpu, r)
res_r = np.empty_like(r)

NLM2_Colour_GPU(r_gpu, g_gpu, b_gpu, img_r_gpu, img_g_gpu, img_b_gpu,np.intc(width),np.intc(height), np.float32(Noise), \
         np.float32(lerpC), block=(nb_ThreadsX,nb_ThreadsY,1), grid=(nb_blocksX,nb_blocksY))

drv.memcpy_dtoh(res_r, r_gpu)
drv.memcpy_dtoh(res_g, g_gpu)
drv.memcpy_dtoh(res_b, b_gpu)
r_gpu.free()
g_gpu.free()
b_gpu.free()
img_r_gpu.free()
img_g_gpu.free()
img_b_gpu.free()

image_brut_GPU=cv2.merge((res_r,res_g,res_b))

tps_GPU_vegetables = time.time() - tps1

print("PyCuda treatment OK")
print ("GPU treatment time : ",tps_GPU_vegetables)
print("")

cv2.imwrite(Out_File_PyCuda, cv2.cvtColor(image_brut_GPU, cv2.COLOR_BGR2RGB), [int(cv2.IMWRITE_JPEG_QUALITY), 95]) # JPEG file format


Ecart1_vegetables = tps_CPU_vegetables/tps_GPU_vegetables
print ("Acceleration factor CPU/GPU : ",Ecart1_vegetables)
```

```python
import time
import cv2
import numpy as np
import pycuda.driver as drv
import pycuda.tools
import pycuda.autoinit
from pycuda.compiler import SourceModule
import pycuda.gpuarray as gpuarray

# Define target sizes
sizes = [(16, 16), (32, 32), (64, 64), (128, 128), (256, 256), (512, 512)]

# Load the original image
original_img = cv2.imread('bird.jpg', cv2.IMREAD_COLOR)
mean, stddev = 0, 20  # Standard deviation for noise level

# Initialize a list to store times
cpu_times = []
gpu_times = []

for size in sizes:
    # Resize the original image
    img = cv2.resize(original_img, size, interpolation=cv2.INTER_AREA)

    # Generate Gaussian noise and add it to the resized image
    gaussian_noise = np.random.normal(mean, stddev, img.shape).astype(np.float32)
    noisy_img = img.astype(np.float32) + gaussian_noise
    noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)
    cv2.imwrite('noisy_image.jpg', noisy_img)


    mod = SourceModule("""
    __global__ void NLM2_Colour(unsigned char *dest_r, unsigned char *dest_g, unsigned char *dest_b, unsigned char *img_r, unsigned char *img_g,
    unsigned char *img_b, int imageW, int imageH, float Noise, float lerpC)
    {

    #define NLM_WINDOW_RADIUS   3
    #define NLM_BLOCK_RADIUS    3

    #define NLM_WEIGHT_THRESHOLD    0.00039f
    #define NLM_LERP_THRESHOLD      0.10f

    __shared__ float fWeights[64];

    const float NLM_WINDOW_AREA = (2.0 * NLM_WINDOW_RADIUS + 1.0) * (2.0 * NLM_WINDOW_RADIUS + 1.0) ;
    const float INV_NLM_WINDOW_AREA = (1.0 / NLM_WINDOW_AREA);

    const long int   ix = blockDim.x * blockIdx.x + threadIdx.x;
    const long int   iy = blockDim.y * blockIdx.y + threadIdx.y;

    const float  x = (float)ix  + 1.0f;
    const float  y = (float)iy  + 1.0f;
    const float cx = blockDim.x * blockIdx.x + NLM_WINDOW_RADIUS + 1.0f;
    const float cy = blockDim.x * blockIdx.y + NLM_WINDOW_RADIUS + 1.0f;
    const float limxmin = NLM_BLOCK_RADIUS + 2;
    const float limxmax = imageW - NLM_BLOCK_RADIUS - 2;
    const float limymin = NLM_BLOCK_RADIUS + 2;
    const float limymax = imageH - NLM_BLOCK_RADIUS - 2;


    long int index4;
    long int index5;

    if(ix>limxmin && ix<limxmax && iy>limymin && iy<limymax){
        //Find color distance from current texel to the center of NLM window
        float weight = 0;

        for(float n = -NLM_BLOCK_RADIUS; n <= NLM_BLOCK_RADIUS; n++)
            for(float m = -NLM_BLOCK_RADIUS; m <= NLM_BLOCK_RADIUS; m++) {
                long int index1 = cx + m + (cy + n) * imageW;
                long int index2 = x + m + (y + n) * imageW;
                weight += ((img_r[index2] - img_r[index1]) * (img_r[index2] - img_r[index1])
                + (img_g[index2] - img_g[index1]) * (img_g[index2] - img_g[index1])
                + (img_b[index2] - img_b[index1]) * (img_b[index2] - img_b[index1])) / (256.0 * 256.0);
                }

        //Geometric distance from current texel to the center of NLM window
        float dist =
            (threadIdx.x - NLM_WINDOW_RADIUS) * (threadIdx.x - NLM_WINDOW_RADIUS) +
            (threadIdx.y - NLM_WINDOW_RADIUS) * (threadIdx.y - NLM_WINDOW_RADIUS);

        //Derive final weight from color and geometric distance
        weight = __expf(-(weight * Noise + dist * INV_NLM_WINDOW_AREA));

        //Write the result to shared memory
        fWeights[threadIdx.y * 8 + threadIdx.x] = weight / 256.0;
        //Wait until all the weights are ready
        __syncthreads();


        //Normalized counter for the NLM weight threshold
        float fCount = 0;
        //Total sum of pixel weights
        float sumWeights = 0;
        //Result accumulator
        float3 clr = {0.0, 0.0, 0.0};

        int idx = 0;

        //Cycle through NLM window, surrounding (x, y) texel

        for(float i = -NLM_WINDOW_RADIUS; i <= NLM_WINDOW_RADIUS + 1; i++)
            for(float j = -NLM_WINDOW_RADIUS; j <= NLM_WINDOW_RADIUS + 1; j++)
            {
                //Load precomputed weight
                float weightIJ = fWeights[idx++];

                //Accumulate (x + j, y + i) texel color with computed weight
                float3 clrIJ ; // Ligne code modifiÃ©e
                int index3 = x + j + (y + i) * imageW;
                clrIJ.x = img_r[index3];
                clrIJ.y = img_g[index3];
                clrIJ.z = img_b[index3];

                clr.x += clrIJ.x * weightIJ;
                clr.y += clrIJ.y * weightIJ;
                clr.z += clrIJ.z * weightIJ;

                //Sum of weights for color normalization to [0..1] range
                sumWeights  += weightIJ;

                //Update weight counter, if NLM weight for current window texel
                //exceeds the weight threshold
                fCount += (weightIJ > NLM_WEIGHT_THRESHOLD) ? INV_NLM_WINDOW_AREA : 0;
            }

        //Normalize result color by sum of weights
        sumWeights = 0.0039f / sumWeights;
        clr.x *= sumWeights;
        clr.y *= sumWeights;
        clr.z *= sumWeights;

        //Choose LERP quotent basing on how many texels
        //within the NLM window exceeded the weight threshold
        float lerpQ = (fCount > NLM_LERP_THRESHOLD) ? lerpC : 1.0f - lerpC;

        //Write final result to global memory
        float3 clr00 = {0.0, 0.0, 0.0};
        index4 = x + (y * imageW);
        index5 = imageW * (iy + 1) + ix + 1;

        clr00.x = img_r[index4] / 256.0;
        clr00.y = img_g[index4] / 256.0;
        clr00.z = img_b[index4] / 256.0;

        clr.x = clr.x + (clr00.x - clr.x) * lerpQ;
        clr.y = clr.y + (clr00.y - clr.y) * lerpQ;
        clr.z = clr.z + (clr00.z - clr.z) * lerpQ;


        dest_r[index5] = (int)(clr.x * 256.0);
        dest_g[index5] = (int)(clr.y * 256.0);
        dest_b[index5] = (int)(clr.z * 256.0);
       }
    }
    """)
    NLM2_Colour_GPU = mod.get_function("NLM2_Colour")
    # Read the noisy image for processing
    image_brut_CV = cv2.imread('noisy_image.jpg', cv2.IMREAD_COLOR)
    height, width, layers = image_brut_CV.shape
    nb_pixels = height * width
    # Set NLM parameters and determine block/grid sizes
    nb_ThreadsX = 8
    nb_ThreadsY = 8
    nb_blocksX = (width // nb_ThreadsX) + 1
    nb_blocksY = (height // nb_ThreadsY) + 1
    # Set NLM parameters
    NLM_Noise = 1.45
    Noise = 1.0/(NLM_Noise*NLM_Noise)
    lerpC = 0.2

    # Algorithm CPU using OpenCV fastNlMeansDenoisingColored routine
    tps1 = time.time()
    param=21.0
    image_brut_CPU=cv2.fastNlMeansDenoisingColored(image_brut_CV, None, param, param, 3, 5) # application filtre denoise software colour
    image_brut_CPU=cv2.cvtColor(image_brut_CPU, cv2.COLOR_BGR2RGB)
    tps_CPU = time.time() - tps1
    print("OpenCV treatment OK")
    print ("CPU treatment time : ",tps_CPU)
    print("")
    cpu_times.append(tps_CPU)

    # GPU processing with PyCuda
    tps1 = time.time()
    b, g, r = cv2.split(image_brut_CV)
    b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
    img_b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
    drv.memcpy_htod(b_gpu, b)
    drv.memcpy_htod(img_b_gpu, b)
    res_b = np.empty_like(b)

    g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
    drv.memcpy_htod(g_gpu, g)
    img_g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
    drv.memcpy_htod(img_g_gpu, g)
    res_g = np.empty_like(g)

    r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
    drv.memcpy_htod(r_gpu, r)
    img_r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
    drv.memcpy_htod(img_r_gpu, r)
    res_r = np.empty_like(r)


    NLM2_Colour_GPU(r_gpu, g_gpu, b_gpu, img_r_gpu, img_g_gpu, img_b_gpu,np.intc(width),np.intc(height), np.float32(Noise), \
         np.float32(lerpC), block=(nb_ThreadsX,nb_ThreadsY,1), grid=(nb_blocksX,nb_blocksY))

    drv.memcpy_dtoh(res_r, r_gpu)
    drv.memcpy_dtoh(res_g, g_gpu)
    drv.memcpy_dtoh(res_b, b_gpu)
    r_gpu.free()
    g_gpu.free()
    b_gpu.free()
    img_r_gpu.free()
    img_g_gpu.free()
    img_b_gpu.free()

    image_brut_GPU=cv2.merge((res_r,res_g,res_b))

    tps_GPU = time.time() - tps1

    gpu_times.append(tps_GPU)

    # # Cleanup GPU memory
    # r_gpu.free()
    # g_gpu.free()
    # b_gpu.free()
    # img_r_gpu.free()
    # img_g_gpu.free()
    # img_b_gpu.free()

# Display CPU and GPU times for each size
for i, size in enumerate(sizes):
    print(f"Size: {size}, CPU Time: {cpu_times[i]:} s, GPU Time: {gpu_times[i]:} s")
    print(f"Acceleration Factor (CPU/GPU) for {size}: {cpu_times[i] / gpu_times[i]:}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data from the previous loop
sizes = [(16, 16), (32, 32), (64, 64), (128, 128), (256, 256), (512, 512)]


# Convert sizes to total pixels for easier plotting
total_pixels = [w * h for w, h in sizes]

cpu_times_ms = [time * 1000 for time in cpu_times]
gpu_times_ms = [time * 1000 for time in gpu_times]

# Create a DataFrame to hold sizes and times
df = pd.DataFrame({
    "Image Size (Pixels)": total_pixels,
    "CPU Time (ms)": cpu_times,
    "GPU Time (ms)": gpu_times,
    "CPU/GPU time": [cpu / gpu for cpu, gpu in zip(cpu_times, gpu_times)]
})

# Plotting CPU and GPU times against image size
plt.figure(figsize=(10, 6))
plt.plot(total_pixels, cpu_times, label='CPU Time (ms)', marker='o', color='b')
plt.plot(total_pixels, gpu_times, label='GPU Time (ms)', marker='o', color='r')

# Adding labels and legend
plt.xlabel("Image Size (Pixels)")
plt.ylabel("Processing Time (ms)")
plt.title("CPU vs GPU Processing Time by Image Size")
plt.legend()
plt.grid(True)
plt.show()

# Display the table as an image if needed
df.style.format({"CPU Time (ms)": "{:.4f}", "GPU Time (ms)": "{:.7f}", "CPU/GPU time": "{:.4f}"})
```

```python
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import time
import cv2
import numpy as np
import pycuda.driver as drv
import pycuda.autoinit
from pycuda.compiler import SourceModule
import matplotlib.pyplot as plt

# Define noise levels
noise_levels = [10, 20, 30]  # Standard deviation for Gaussian noise

# SSIM and PSNR results
results = []

# Load original image
original_img = cv2.imread('bird.jpg', cv2.IMREAD_COLOR)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

for noise_std in noise_levels:
    # Generate Gaussian noise
    gaussian_noise = np.random.normal(0, noise_std, original_img.shape).astype(np.float32)

    # Add noise
    noisy_img = original_img.astype(np.float32) + gaussian_noise
    noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

    # Apply CPU Denoise
    start_time = time.time()
    denoised_img_cpu = cv2.fastNlMeansDenoisingColored(noisy_img, None, 21, 21, 7, 21)
    cpu_time = time.time() - start_time

    # Apply GPU Denoise (using the provided CUDA kernel function)
    # Assuming `NLM2_Colour_GPU` and necessary setup from above code

    # Set up GPU buffers
    b, g, r = cv2.split(noisy_img)
    b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
    g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
    r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)
    img_b_gpu = drv.mem_alloc(b.size * b.dtype.itemsize)
    img_g_gpu = drv.mem_alloc(g.size * g.dtype.itemsize)
    img_r_gpu = drv.mem_alloc(r.size * r.dtype.itemsize)

    drv.memcpy_htod(b_gpu, b)
    drv.memcpy_htod(g_gpu, g)
    drv.memcpy_htod(r_gpu, r)
    drv.memcpy_htod(img_b_gpu, b)
    drv.memcpy_htod(img_g_gpu, g)
    drv.memcpy_htod(img_r_gpu, r)

    # Denoise on GPU
    start_time = time.time()
    NLM2_Colour_GPU(r_gpu, g_gpu, b_gpu, img_r_gpu, img_g_gpu, img_b_gpu,
                    np.intc(original_img.shape[1]), np.intc(original_img.shape[0]),
                    np.float32(1.0 / (1.45 * 1.45)), np.float32(0.2),
                    block=(8, 8, 1), grid=(original_img.shape[1] // 8 + 1, original_img.shape[0] // 8 + 1))
    gpu_time = time.time() - start_time

    # Retrieve GPU result
    res_b = np.empty_like(b)
    res_g = np.empty_like(g)
    res_r = np.empty_like(r)
    drv.memcpy_dtoh(res_r, r_gpu)
    drv.memcpy_dtoh(res_g, g_gpu)
    drv.memcpy_dtoh(res_b, b_gpu)

    # Merge GPU channels
    denoised_img_gpu = cv2.merge((res_r, res_g, res_b))

    # Calculate SSIM and PSNR
    ssim_cpu = ssim(original_img, denoised_img_cpu, multichannel=True, win_size=7, channel_axis=2)
    psnr_cpu = psnr(original_img, denoised_img_cpu)
    ssim_gpu = ssim(original_img, denoised_img_gpu, multichannel=True, win_size=7, channel_axis=2)
    psnr_gpu = psnr(original_img, denoised_img_gpu)

    # Append results
    results.append({
        'Noise Level (stddev)': noise_std,
        'CPU SSIM': ssim_cpu,
        'GPU SSIM': ssim_gpu,
        'CPU PSNR': psnr_cpu,
        'GPU PSNR': psnr_gpu,
        'CPU Time (s)': cpu_time,
        'GPU Time (s)': gpu_time
    })

# Display results
import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)
```
